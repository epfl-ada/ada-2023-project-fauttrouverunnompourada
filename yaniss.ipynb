```python


import pandas as pd
import json

# Exemple de création de DataFrame (à remplacer par votre DataFrame réel)


# Créer une copie du DataFrame
new_movie_data = movie_data.copy()


s = 0


def extract_genre(genre_str):
    global s  # Indique que la variable s fait référence à la variable déclarée en dehors de la fonction
    try:
        genre_dict = json.loads(genre_str.replace("'", '"'))
        return list(genre_dict.values())
    except json.JSONDecodeError:
        s += 1
        # print(f"Error converting string to dictionary. JSONDecodeError. s = {s}")
        return None

print(s)


# Appliquer la fonction à la colonne "Genres" de la nouvelle table
new_movie_data['Genres'] = new_movie_data['Genres'].apply(extract_genre)

new_movie_data

# Compter le nombre de valeurs NaN ou None par colonne
nan_count = new_movie_data.isna().sum()

# Afficher le résultat
print("Nombre de valeurs NaN ou None par colonne :")
print(nan_count)


```
